```python
class Solution:
    def isValid(self, s: str) -> bool:
        bracket={'{':'}' , '(':')' , '[':']' }
        stack=[]
        for i in s:
            if i in bracket:
                stack.append(i)
            else:
                if stack and bracket[stack[-1]]==i:
                    stack.pop()
                else:
                    return False
        if stack:
            return False
        return True
```

# Valid Parentheses Using Stack

This algorithm checks whether a given string containing only parentheses (`()`, `{}`, `[]`) is valid, meaning:
- Every opening bracket has a corresponding closing bracket.
- The brackets are properly nested.

## How It Works
1. **Initialization**:  
   - Use a dictionary (`bracket`) to map opening brackets to their corresponding closing brackets.
   - Use a stack (`stack`) to keep track of open brackets.

2. **Traversal & Validation**:  
   - Iterate through each character in the string.
   - If it's an **opening bracket**, push it onto the stack.
   - If it's a **closing bracket**:
     - Check if the stack is **not empty** and the top of the stack matches the expected closing bracket.
     - If it matches, pop the stack.
     - Otherwise, return `False` (invalid sequence).

3. **Final Check**:  
   - If the stack is **empty**, the string is valid.
   - If the stack **contains leftover brackets**, return `False`.

## Corner Cases
- **Empty String**: Considered valid (`True`).
- **Single Bracket**: Always invalid (`False`).
- **Mixed Unmatched Brackets**: Detected during traversal.
- **Correctly Nested Brackets**: Valid and returns `True`.

## Complexity
- **Time Complexity**: `O(n)`, as each character is processed once.
- **Space Complexity**: `O(n)`, since the stack stores at most `n/2` elements in the worst case.
