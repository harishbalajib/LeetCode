

[ **PROBLEM LINK**  ](https://leetcode.com/problems/merge-two-sorted-lists/description/)

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def mergeTwoLists(self, list1, list2) :
        if list1 is None:
            return list2
        if list2 is None:
            return list1
        if list1.val <= list2.val:
            list1.next = self.mergeTwoLists(list1.next, list2)
            return list1
        else:
            list2.next = self.mergeTwoLists(list1, list2.next)
            return list2
```

# Recursive Approach to Merge Two Sorted Linked Lists

This algorithm merges two **sorted singly linked lists** into a single sorted linked list using recursion.

## **How It Works**
1. **Base Cases**:  
   - If `list1` is `None`, return `list2` (remaining nodes).
   - If `list2` is `None`, return `list1` (remaining nodes).

2. **Recursive Merging**:  
   - Compare `list1.val` and `list2.val`.
   - Attach the smaller value node to the merged list.
   - Recursively call `mergeTwoLists()` on the next node of the chosen list.

3. **Return the Head**:  
   - The function returns the head of the newly merged list.

## **Corner Cases**
- **Both Lists Empty**: Returns `None`.
- **One List Empty**: Returns the non-empty list.
- **Lists of Unequal Length**: Works correctly by handling nodes one at a time.

## **Complexity**
- **Time Complexity**: `O(n + m)`, where `n` and `m` are the lengths of the two lists (each node is processed once).
- **Space Complexity**: `O(n + m)`, due to recursive function calls stored in the call stack.

# Much More Optimized Solution

```python

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeTwoLists(self, list1: ListNode, list2: ListNode) -> ListNode:
        dummy = node = ListNode()

        while list1 and list2:
            if list1.val < list2.val:
                node.next = list1
                list1 = list1.next
            else:
                node.next = list2
                list2 = list2.next
            node = node.next

        node.next = list1 or list2

        return dummy.next
```

# Iterative Approach to Merge Two Sorted Linked Lists

This algorithm merges two **sorted singly linked lists** into a single sorted linked list using an iterative approach.

## **How It Works**
1. **Initialize a Dummy Node**:  
   - A `dummy` node is used as a placeholder to simplify edge cases.
   - A `node` pointer is set to `dummy` to keep track of the merged list.

2. **Iterate Through Both Lists**:  
   - Compare `list1.val` and `list2.val`.
   - Attach the smaller node to `node.next` and move the pointer forward.
   - Move the corresponding list (`list1` or `list2`) to its next node.

3. **Attach Remaining Nodes**:  
   - If one list is exhausted, attach the remaining nodes of the other list directly.

4. **Return the Merged List**:  
   - Since `dummy.next` points to the merged list, return `dummy.next`.

## **Corner Cases**
- **Both Lists Empty**: Returns `None`.
- **One List Empty**: Returns the non-empty list.
- **Lists of Unequal Length**: Correctly appends remaining elements.

## **Complexity**
- **Time Complexity**: `O(n + m)`, where `n` and `m` are the lengths of the two lists.
- **Space Complexity**: `O(1)`, since merging is done in place without extra memory.
