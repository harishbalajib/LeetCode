```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """

        count=[0]*3
        for i in nums:
            count[i]+=1
        n=len(count)
        i=0
        for idx in range(n):
            for _ in range(count[idx]):
                nums[i]=idx
                i+=1
        

        
```

# **Sort Colors (Counting Sort Approach)**

This algorithm sorts an array containing only `0`, `1`, and `2` in-place using **counting sort**.

## **How It Works**
1. **Count the Occurrences**:
   - Create a `count` array of size `3` to store the frequency of `0`, `1`, and `2`.
   - Iterate over `nums`, incrementing the respective index in `count`.

2. **Reconstruct the Array**:
   - Iterate over the `count` array.
   - Place `0`s, `1`s, and `2`s back into `nums` based on their frequency.

---

## **Complexity Analysis**
| Operation | Time Complexity |
|-----------|----------------|
| **Counting Occurrences** | **O(n)** |
| **Reconstructing the Array** | **O(n)** |
| **Total Complexity** | **O(n)** |
| **Space Complexity** | **O(1)** (uses only a fixed-size `count` array) |

---

## **Why This Works Well?**
- **In-Place**: It doesn't use extra space beyond the fixed `count` array.
- **Optimal for Small Range Values (`0-2`)**: Since the range is limited, counting sort is **fast and efficient**.

---

## **Corner Cases**
- **All Same Numbers**: Works efficiently (e.g., `[0,0,0]`, `[2,2,2]`).
- **Already Sorted Array**: Still performs `O(n)`, but is efficient.
- **Reverse Sorted Array**: Still sorts in `O(n)` without extra space.

---
