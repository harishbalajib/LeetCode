```python
# The isBadVersion API is already defined for you.
# def isBadVersion(version: int) -> bool:

class Solution:
    def firstBadVersion(self, n: int) -> int:
        left=1
        right=n
       
        while left < right:

            bad=(left+right)//2

            if isBadVersion(bad):
                right=bad

            else:
                left=bad+1

        return left
```

# **First Bad Version (Binary Search Approach)**

This algorithm efficiently finds the **first bad version** in a sequence of versions using **binary search**.

## **How It Works**
1. **Define Search Range**:
   - Start with `left = 1` and `right = n` (searching within versions `1` to `n`).

2. **Binary Search for the First Bad Version**:
   - Compute `mid = (left + right) // 2`.
   - Use `isBadVersion(mid)` to check if `mid` is a **bad version**:
     - If `mid` is bad, search **left half** (`right = mid`).
     - If `mid` is good, search **right half** (`left = mid + 1`).

3. **Return the First Bad Version**:
   - The loop stops when `left == right`, pointing to the first bad version.

---

## **Complexity Analysis**
| Operation | Time Complexity |
|-----------|----------------|
| **Binary Search Range (`1` to `n`)** | **O(log n)** |

- **Binary search** reduces the search space logarithmically, leading to **O(log n) complexity**.
- More efficient than a **brute-force approach (`O(n)`)**, which checks every version one by one.

---

## **Why This Works Well?**
- **Efficient for large `n`**: Works even when `n` is **very large (`10`)**.
- **Minimizes API Calls**: Uses **logarithmic API calls** instead of checking each version sequentially.

---

## **Corner Cases**
- **All Versions Are Good (`isBadVersion(n) = False`)**: Should not happen per problem constraints.
- **First Version is Bad (`isBadVersion(1) = True`)**: Algorithm correctly returns `1`.
- **Last Version is Bad (`isBadVersion(n) = True`)**: Finds the first occurrence efficiently.

---

## **Final Notes**
- **Binary search ensures the earliest bad version is found efficiently**.
- **Reduces API calls compared to a naive linear search** (`O(n)`).
- **Ideal for large-scale version control problems**.
