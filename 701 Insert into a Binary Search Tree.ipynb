```python
# Definition for a binary tree node.
from typing import Optional
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def insertIntoBST(self, root: Optional[TreeNode], val: int) -> Optional[TreeNode]:
        parser=root

        while parser :

            if parser.val>val:
                if parser.left:
                    parser=parser.left
                else:
                    parser.left=TreeNode(val)
                    return root

            elif parser.val<val:
                if parser.right:
                    parser=parser.right
                else:
                    parser.right=TreeNode(val)
                    return root

        return TreeNode(val)
        
        
```

# **Insert into a Binary Search Tree (Iterative Approach)**

This algorithm inserts a new node with value `val` into a **Binary Search Tree (BST)** while maintaining the BST properties.

---

## **How It Works**
1. **Handle Empty Tree (`root == None`)**:
   - If the BST is empty, return a new node with `val`.

2. **Traverse the BST**:
   - Start from `root`, using `parser` as a pointer.
   - If `val < parser.val`, move **left** (`parser = parser.left`).
   - If `val > parser.val`, move **right** (`parser = parser.right`).

3. **Insert the New Node**:
   - When we find an **empty position (`None`)**, create a new node and attach it as **left or right child**.
   - Return `root` to maintain the tree structure.

---

## **Complexity Analysis**
| Operation | Time Complexity | Space Complexity |
|-----------|----------------|------------------|
| **Best Case (Insert at Root)** | **O(1)** | **O(1)** |
| **Average Case (Balanced BST)** | **O(log n)** | **O(1)** |
| **Worst Case (Skewed Tree, like a Linked List)** | **O(n)** | **O(1)** |

- **Balanced BST (`O(log n)`)**: Each step halves the search space.
- **Skewed BST (`O(n)`)**: If the tree is **highly unbalanced**, it degenerates into a linked list.

---

## **Why Use an Iterative Approach?**
- **Avoids Recursion Overhead**: No extra recursive calls, reducing stack space usage.
- **More Memory Efficient (`O(1)`)**: Uses only a pointer (`parser`), avoiding recursive function call stacks (`O(log n)`).
- **Faster Execution in Large Trees**: No recursive function calls help with performance.

---

## **Corner Cases**
- **Empty Tree (`root = None`)**: Creates a new tree with the given `val`.
- **Insert at Leaf Node**: Works correctly at the correct BST position.
- **Insert in a Skewed Tree**: Runs in `O(n)`, but still correctly places the node.

---

## **Final Notes**
- **Efficient iterative insertion with `O(log n)` complexity for balanced BSTs**.
- **Best suited for large BSTs where recursion depth might be a concern**.
- **Maintains BST properties while efficiently inserting new values**.
