

[ **PROBLEM LINK**  ](https://leetcode.com/problems/merge-k-sorted-lists/)

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:    
    def mergeKLists(self, lists):
        if not lists:
            return None
        
        while len(lists)>1:
            mergedLists=[]

            for i in range(0,len(lists),2):
                list1=lists[i]
                list2=lists[i+1] if (i+1) < len(lists) else None
                mergedLists.append(self.mergeList(list1,list2))
            lists=mergedLists

        return lists[0]

    def mergeList(self,head1,head2):

        dummy=final=ListNode()

        while head1 and head2:

            if head1.val<=head2.val:
                dummy.next=head1
                head1=head1.next
                dummy=dummy.next

            else:
                dummy.next=head2
                head2=head2.next
                dummy=dummy.next

        if head1:
            dummy.next=head1

        else:
            dummy.next=head2
            
        return final.next

```

# Merging K Sorted Linked Lists (Divide and Conquer)

This algorithm merges **K sorted linked lists** into a single sorted linked list using the **divide and conquer approach**.

## **How It Works**
### **1. Base Case**
- If `lists` is empty, return `None` (no lists to merge).

### **2. Pairwise Merging (Divide and Conquer)**
- While `lists` contains more than one linked list:
  - Initialize an empty `mergedLists` list.
  - Merge pairs of lists using `mergeList(head1, head2)`.
  - Store merged results in `mergedLists`, reducing the number of lists by half in each iteration.
  - Continue until only one merged list remains.

### **3. Merging Two Lists (`mergeList`)**
- Uses a **dummy node** to simplify list construction.
- Iterates through both lists:
  - Adds the smaller node to the merged list.
  - Moves the pointer forward in the list from which the node was taken.
- Appends any remaining nodes if one list is exhausted.

---

## **Complexity Analysis**
| Scenario | Complexity |
|----------|------------|
| **Merging Two Lists (`mergeList`)** | **O(N + M)** (where `N` and `M` are the lengths of two lists) |
| **Total Complexity (Divide and Conquer Merging)** | **O(N log K)** (where `N` is total nodes and `K` is the number of lists) |

### **Why O(N log K)?**
- **Merging two lists takes O(N)**.
- **Each merge step halves the number of lists**, leading to **O(log K) levels**.
- **Total operations: O(N log K)**.

---

## **Corner Cases**
- **Empty Input (`lists = []`)**: Returns `None`.
- **Single List (`lists = [head]`)**: Returns `head` unchanged.
- **Lists of Varying Lengths**: Works correctly due to `mergeList()` handling different list sizes.

---

## **Final Notes**
- **Efficient alternative to the naive O(KN) merging** (merging sequentially).
- **Optimized compared to min-heap-based priority queue O(N log K)** in some cases.
- **Best suited for merging a large number of sorted linked lists efficiently**.
