# Remove Element Using Two-Pointer Swap

This algorithm removes all occurrences of a target value from an array using two pointers:
- **Left Pointer (`left`)**: Iterates from the beginning to identify target elements.
- **Right Pointer (`right`)**: Points to the end and is used for swapping when a target value is found.

## How It Works
1. **Initialization**:  
   - `left` starts at index `0`, and `right` starts at the last index.
2. **Traversal & Swap**:  
   - If `nums[left]` equals `val`, it is swapped with `nums[right]`, and `right` is decremented.
   - If `nums[left]` is not `val`, simply increment `left`.
3. **Result**:  
   - The final value of `left` represents the number of elements not equal to `val`.

## Corner Cases
- **Empty Array**: Returns 0 immediately.
- **All Elements Are the Target**: The final length is 0 after processing.
- **No Target Elements**: The array remains unchanged, and the returned length is the original array size.

## Complexity
- **Time Complexity**: O(n), since each element is processed at most once.
- **Space Complexity**: O(1), as all operations are done in-place.


```python
class Solution:
    def removeElement(self, nums, val):
        left, right = 0, len(nums) - 1
        while left <= right:
            if nums[left] == val:
                nums[left], nums[right] = nums[right], nums[left]
                right -= 1
            else:
                left += 1
        return left
```

# Remove Element Using Shift-Based Approach

This algorithm removes all occurrences of a target value (`val`) from an array by shifting non-target elements forward.

## How It Works
1. **Initialization**:  
   - Use a single pointer (`left`) starting at index `0` to keep track of the new position for non-target elements.

2. **Traversal & Shift**:  
   - Iterate over the array with another pointer (`right`).
   - If `nums[right]` is **not** `val`, copy it to `nums[left]` and increment `left`.
   - If `nums[right]` **is** `val`, skip it (do nothing).

3. **Result**:  
   - The final value of `left` represents the number of elements that are not `val`.
   - The array is modified **in-place**, with the first `left` elements containing the required values.

## Corner Cases
- **Empty Array**: Returns `0` immediately.
- **All Elements Are the Target**: The final length is `0` after processing.
- **No Target Elements**: The array remains unchanged, and the returned length is the original array size.

## Complexity
- **Time Complexity**: `O(n)`, since each element is processed exactly once.
- **Space Complexity**: `O(1)`, as all operations are done in-place.


```python
from typing import List
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        left,right=0,0
        n=len(nums)
        for right in range(n):
            if nums[right]!=val:
                nums[left]=nums[right]
                left+=1
        return left
```
