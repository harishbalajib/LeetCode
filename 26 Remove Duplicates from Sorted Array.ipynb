# Remove Duplicates Using Two Pointers

This algorithm uses two pointers:
- **Insertion Pointer (`insert_idx`)**: Marks the position where the next unique element should be placed.
- **Traversal Pointer (`pointer`)**: Iterates through the array to identify unique elements.

## How It Works
1. **Initialization**:  
   - Set `insert_idx` to 1 since the first element is always unique.
2. **Traversal**:  
   - Start from the second element (`pointer = 1`).
   - If the current element differs from the previous one, it is unique.  
   - Place it at `nums[insert_idx]` and increment `insert_idx`.
3. **Result**:  
   - `insert_idx` holds the new length of the array with unique elements.

## Corner Cases
- **Empty Array**: Returns 0.
- **Single Element**: Returns 1.
- **All Elements are Duplicates**: Results in an array with one element.
- **No Duplicates**: The array remains unchanged, returning its original length.

## Complexity
- **Time Complexity**: O(n), where n is the number of elements (single pass through the array).
- **Space Complexity**: O(1), as the operations are performed in-place.

```python
class Solution:
    def removeDuplicates(self, nums):
        insert_idx=1
        n=len(nums)
        for pointer in range(1,n):
            if nums[pointer]!=nums[pointer-1]:
                nums[insert_idx]=nums[pointer]
                insert_idx+=1
            
        return insert_idx
                
```
