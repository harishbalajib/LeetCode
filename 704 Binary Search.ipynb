```python
from typing import List
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        left =0
        right=len(nums)-1
        while left<=right:
            mid=(left+right)//2
            if nums[mid]<target:
                left=mid+1
            elif nums[mid]>target:
                right=mid-1
            else:
                return mid
        return -1
        
```

# **Binary Search in a Sorted Array**

This algorithm implements **binary search** to find a target value in a sorted array.

## **How It Works**
1. **Initialize Pointers**:
   - `left = 0` (start of the array).
   - `right = len(nums) - 1` (end of the array).

2. **Binary Search Loop**:
   - Compute `mid = (left + right) // 2`.
   - If `nums[mid] == target`, return `mid` (target found).
   - If `nums[mid] < target`, search the **right half** (`left = mid + 1`).
   - If `nums[mid] > target`, search the **left half** (`right = mid - 1`).

3. **Return `-1` if Target is Not Found**.

---

## **Complexity Analysis**
| Operation | Time Complexity |
|-----------|----------------|
| **Binary Search** | **O(log n)** |

- The search space **halves** in each iteration, leading to **logarithmic time complexity**.

---

## **Why is This Efficient?**
- **Faster than Linear Search (`O(n)`)**: Works well for large sorted arrays.
- **Minimal Space Usage (`O(1)`)**: Only a few extra variables are used.

---

## **Corner Cases**
- **Empty Array (`nums = []`)**: Returns `-1`.
- **Target at First or Last Index**: Successfully finds it in `O(1)`.
- **Target Not in Array**: Correctly returns `-1`.
- **Array of One Element (`nums = [x]`)**: Works correctly.

---

## **Alternative Approach (Recursive Binary Search)**
- Instead of a loop, use **recursion** to divide the search space.
- Similar **O(log n)** complexity, but with **O(log n) space** due to recursive calls.
