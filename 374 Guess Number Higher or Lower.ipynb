```python
# The guess API is already defined for you.
# @param num, your guess
# @return -1 if num is higher than the picked number
#          1 if num is lower than the picked number
#          otherwise return 0
# def guess(num: int) -> int:

class Solution:
    def guessNumber(self, n: int) -> int:
        low=1
        high=n
        while low<=high:
            mid=(low+high)//2
            pred=guess(mid)
            if pred==1:
                low=mid+1
            elif pred==-1:
                high=mid-1
            else:
                return mid
            
        
```

# **Guess Number Higher or Lower (Binary Search Approach)**

This algorithm efficiently finds the target number using **binary search**.

## **How It Works**
1. **Initialize Search Space**:
   - `low = 1`, `high = n` (search range from `1` to `n`).

2. **Binary Search Loop**:
   - Compute `mid = (low + high) // 2`.
   - Use `guess(mid)`, which:
     - Returns `1` if `mid` is too low.
     - Returns `-1` if `mid` is too high.
     - Returns `0` if `mid` is correct.

3. **Adjust Search Range**:
   - If `guess(mid) == 1`: Move `low = mid + 1` (search right).
   - If `guess(mid) == -1`: Move `high = mid - 1` (search left).
   - If `guess(mid) == 0`: Return `mid` (correct answer).

---

## **Complexity Analysis**
| Operation | Time Complexity |
|-----------|----------------|
| **Binary Search** | **O(log n)** |

- Each iteration halves the search range, leading to **logarithmic complexity**.

---

## **Why This Works Well?**
- **Efficient for Large `n`**: Even for `n = 10⁶`, it runs in about `log(10⁶) ≈ 20` iterations.
- **Minimizes Comparisons**: Instead of checking every number (`O(n)`), binary search narrows down quickly.

---

## **Corner Cases**
- **`n = 1`**: The answer is immediately `1`.
- **`target` at the Edges (`1` or `n`)**: Correctly finds in `O(log n)`.
- **Already Correct at `mid`**: Returns in the first iteration.

---

## **Final Notes**
- **Best suited for a guessing game where we get hints (`guess()` function).**
- **Optimized approach using binary search for efficient number finding.**
