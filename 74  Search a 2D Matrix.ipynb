```python
from typing import List

class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:

        ROWS,COLS=len(matrix),len(matrix[0])
        top,bot=0,ROWS-1
        
        while top<=bot:
            curr_row=(top+bot)//2
            if matrix[curr_row][0]>target:
                bot = curr_row-1
            elif matrix[curr_row][-1]<target:
                top = curr_row+1
            else:
                break
        
        if top>bot:
            return False
        
        curr_row=(top+bot)//2

        left,right=0,COLS-1
        while left<=right:
            mid = (left+right)//2
            if matrix[curr_row][mid]>target:
                right=mid-1
            elif matrix[curr_row][mid]<target:
                left=mid+1
            else:
                return True
        
        return False
```

# **Search in a Sorted 2D Matrix (Binary Search Approach)**

This algorithm efficiently searches for a target value in a **row-wise and column-wise sorted 2D matrix** using **binary search**.

## **How It Works**
1. **Binary Search on Rows**:
   - Find the potential row where the target might exist.
   - If `target < matrix[curr_row][0]`, move **up** (`bot = curr_row - 1`).
   - If `target > matrix[curr_row][-1]`, move **down** (`top = curr_row + 1`).
   - Otherwise, the target is in this row.

2. **Binary Search Within the Row**:
   - Perform **binary search** on the identified row.
   - If `target > mid-element`, search the **right half**.
   - If `target < mid-element`, search the **left half**.
   - If found, return `True`.

3. **Return `False` if Not Found**.

---

## **Complexity Analysis**
| Operation | Time Complexity |
|-----------|----------------|
| **Binary Search on Rows** | **O(log m)** |
| **Binary Search on Selected Row** | **O(log n)** |
| **Total Complexity** | **O(log m + log n) = O(log(mn))** |

- `m` = Number of rows
- `n` = Number of columns

---

## **Why is This Efficient?**
- **Two Binary Searches**: Instead of scanning the entire matrix (`O(mn)`), we perform **only `O(log(mn))`** operations.
- **Optimized for Large Matrices**: Works well for very large `m x n` matrices.

---

## **Corner Cases**
- **Empty Matrix (`matrix = []`)**: Return `False`.
- **Single Row or Single Column Matrices**: Works correctly.
- **Target Not in Matrix**: Correctly returns `False`.
- **Target at the Edges (First/Last Elements)**: Efficiently finds it.

---

## **Alternative Approach (Flattened Binary Search)**
- Instead of **row-wise search**, treat the matrix as a **flattened sorted list**.
- Perform **one single binary search** on index mapping `(row = mid // n, col = mid % n)`.
- Achieves the **same O(log(mn)) complexity**.
